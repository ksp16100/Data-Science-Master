Q4. Explain with an example:#

=>try and else

=>finally

=>raise

```python
#try and else
def multiplay(x,y):
  try:
    result= x*y
  except Exception as e:
    print("this is exception block", e)
  else:
    return  result
```

```python
print(multiplay(10,15))
```

```python
#finally
def add(x,y):
  try:
    result= x+y
  finally :
    return result
```

```python
print(add(7,8))
```

```python
#raise
def divied(x,y):
  if y == 0:
      raise ZeroDivisionError("can not divied by zero")
  return x/y
try :
    print(divied(10,2))
    print(divied(10,0))

except ZeroDivisionError as e:
    print(e)

```
