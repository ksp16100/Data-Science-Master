Q6. Create a custom exception class. Use this class to handle an exception.

```python
class CustomException(Exception):
   pass

def divide(x,y):
    if y == 0:
       raise CustomException("can not divide by zero")
    return x / y

try:
  print(divide(66,3))
  print(divide(15,0))
except CustomException as e:
  print(e)
```
