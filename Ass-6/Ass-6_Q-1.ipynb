Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

=>A class is a blueprint for creating objects providing initial values for state , and implementations of behavior.

An object is an instance of a class, created at runtime. Each object has its own unique set of attributes and methods that operate on that data.

```python
class employ():
  def __init__(self,name,rank,email_id,mobil_num):
    self.name= name
    self.rank= rank
    self.email_id= email_id
    self.mobil_num=mobil_num

  def employ_datels(self):
    return self.name,self.rank,self.email_id,self.mobil_num
```

```python
em = employ("jay","clark","jay@gmail.com",8965231254)
```

```python
em.employ_datels()
```
