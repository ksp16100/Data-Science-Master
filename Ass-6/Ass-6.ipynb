Q5. What is inheritance? Give an example for each type of inheritance.

inheritance is a mechanism in object-oriented programming that allows a new class to be derived from an existing class, inheriting its attributes and behavior. This is useful for creating new classes that are related to existing classes and share some of their functionality.

Types of Inheritance

1) Single Inheritance

Single inheritance enables a derived class to inherit properties from a single parent class, thus enabling code reusability and the addition of new features to existing code.



```python
class parent():
   def class1(self):
    print("This is class 1")

class child(parent):
  def class2(self):
    print("This is class 2")
```

```python
p=parent()
```

```python
p.class1()
```

```python
c= child()
```

```python
c.class2()
```

2) Multiple Inheritance:

When a class can be derived from more than one base class this type of inheritance is called multiple inheritances. In multiple inheritances, all the features of the base classes are inherited into the derived class.



```python
class std1():
  def class1(self):
    print("this is my std 1")
class std2():
  def class2(self):
    print("this is my std 2")
class all_std(std1,std2):
  pass
```

```python
std= all_std()
```

```python
std.class1()
```

```python
std.class2()
```

3) Multilevel Inheritance :

In multilevel inheritance, features of the base class and the derived class are further inherited into the new derived class. This is similar to a relationship representing a child and a grandfather.

```python
class Grandfather():
  def __init__(self,grandfathername):
    self.grandfathername= grandfathername

class Father(Grandfather):
  def __init__(self,fathername,grandfathername):
    self.fathername= fathername
    Grandfather.__init__(self,grandfathername)

class Son(Father):
  def __init__(self,sonname, fathername, grandfathername):
    self.sonname= sonname
    Father.__init__(self, fathername,grandfathername)


  def print_name(self):
      print("grandfather name:-",self.grandfathername)
      print("father name:-",self.fathername)
      print("son name:-",self.sonname)
```

```python
s1 = Son("jay", "vijay", "kalpesh")
```

```python
print(s1.grandfathername)
```

```python
s1.print_name()
```

4) Hierarchical Inheritance:

When more than one derived class are created from a single base this type of inheritance is called hierarchical inheritance. In this program, we have a parent (base) class and two child (derived) classes.

```python
class parent():
  def func1(self):
    print("this is parent")

class child1(parent):
  def func2(self):
    print("this is child 1")

class child2(parent):
  def func3(self):
    print("this is chold 3")
```

```python
class_obj = child1()
```

```python
class_obj2 = child2()
```

```python
class_obj.func1()
```

```python
class_obj.func1()
```

```python
class_obj.func2()
```

```python
class_obj2.func1()
```

```python
class_obj2.func3()
```

5).Hybrid Inheritance:

Inheritance consisting of multiple types of inheritance is called hybrid inheritance.



```python
class college():
  def func1(self):
    print("this is my college")

class bba(college):
  def func2(self):
    print("this is my bba college")

class bcom(college):
  def func3(self):
    print("this is my bcom college")

class bca(bba,college):
  def func3(self):
    print("this is bca collage")
```

```python
obj= bca()
```

```python
obj.func1()
```

```python
obj.func2()
```

```python
obj.func3()
```
