=>args and kwargs are used in functions to allow them to accept a varying number of arguments. *args is used to send a non-keyworded variable length argument list to the function. It allows you to pass a variable number of arguments to a function, which will be received as a tuple. 

```python
def text(*arge):
  return arge
```

=>kwargs is used to pass keyworded variable length of arguments to a function, which will be received as a dictionary. Here's an example of how you can use args and *kwargs in a function:

```python
def text(**kwargs):
  return kwargs
```

```python
text()
```

```python
type(text())
```
