Q1. Which keyword is used to create a function? Create a function to return a list of odd numbers in the
range of 1 to 25.

```python

def odd():
    l  = []
    for i in range(1,26):
        if i % 2 !=0:
          l.append(i)
    return l
odd()
  

```

**kwargs allows us to pass a variable number of keyword arguments to a Python function
